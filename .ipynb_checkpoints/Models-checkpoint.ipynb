### Orange Production Prediction Model
Goal of the project is to predict production volume of Fresh Oranges in EU.

Parameters will be used:
- Utilized agricultural area 
- Agricultural energy consumption 
- Orange Production (area)
- Orange Import Volume
- Orange Export Volume
- Orange Prices per Kg
- Target Value: Orange Production (volume) >> harvested_production column

```python
#importing libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

### Get sorted data from SQL

- Combining data to obtain a Dataset

```python
query1 = "SELECT a.country_code, a.year, a.value AS harvested_area, \
                 e.value AS energy_consumption, p.value AS harvested_production, u.value AS eu_utilized_area \
          FROM area_har_cult_prod a \
          JOIN energy_consumption e ON a.country_code = e.country_code AND a.year = e.year \
          JOIN harvested_prod_eu_stand_humidity p ON e.country_code = p.country_code AND e.year = p.year \
          JOIN eu_utilised_area u ON p.country_code = u.country_code AND p.year = u.year;"
```

```python
country_year_data = pd.read_sql_query(query1, engine)
```

```python
totals = country_year_data.groupby('year').agg({'harvested_area':sum,'energy_consumption':sum, 'harvested_production':sum, 'eu_utilized_area':sum}).reset_index()
totals
```

- Fresh orange import - export

```python
query2 = '''SELECT year, ROUND(AVG(export_volume),2) AS avg_export_volume, ROUND(AVG(import_volume),2) AS avg_import_volume 
FROM orange_import_export_monthly 
GROUP BY year;''' 
```

```python
import_export = pd.read_sql_query(query2, engine)
```

```python
import_export['year'] = import_export['year'].astype(int)
import_export
```

- Orange Prices

```python
query3= '''SELECT year, ROUND(AVG(price),2) AS avg_price FROM orange_prices GROUP BY year;'''
```

```python
prices = pd.read_sql_query(query3, engine)
prices
```

- Merging the data and I will drop 2022 values since there are plenty of missing values

```python
data = pd.merge(pd.merge(import_export, prices, how='outer', on='year'), totals, how='outer', on='year')
data = data[data['year'] != 2022]
data
```

```python
data.describe().T
```

### Modeling

```python
data = data.drop('year', axis=1)
data
```

- I checked the correlation to prevent overfitting of the model, I decided to keep all the variables

```python
correlations_matrix = data.corr() 
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
# X-y Split
y = data['harvested_production']
X = data.drop(['harvested_production'], axis=1)
```

```python
#train-test-split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) # default fraction is .2
```

```python
#scale numericals
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train) 
X_train = transformer.transform(X_train)
X_test = transformer.transform(X_test)
```

- Comparing different models

```python
model1 = DecisionTreeRegressor()
model2 = LinearRegression()
model3 = KNeighborsRegressor()
model4 = RandomForestRegressor()
model5 = MLPRegressor()



model_pipeline = [model1, model2, model3, model4, model5]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN', 'Random Forest Regressor', 'MLP Regressor' ]

val_scores = {}

for model, model_name in zip(model_pipeline,model_names):
    model.fit(X_train, y_train)
    val_scores[model_name] = model.score(X_test,y_test)
print(val_scores)
```

### Results

This was a trial model which will be improved with additional data. Models scores are negative since we have very few data, data is sorted and time sequence was not considered in the modeling.
